```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data as pdr
```

```python
start_date = '2004-01-01'
end_date = '2019-01-01'
marketdata = yf.download("^GSPC", start=start_date, end=end_date)
```

```python
mreturns = marketdata['Adj Close']
mreturns.index = pd.to_datetime(mreturns.index)
mreturns
```

```python
market_last = mreturns.resample('y').last()
```

```python
USB = np.array([
    0.174, 0.170, 0.120, -0.254, 0.265, 0.264, 0.144, 0.206, 0.320, 0.219,
    0.066, 0.109, 0.238, -0.012, 0.306])
UST = np.array([
    0.142, 0.236, 0.062, -0.200, 0.266, 0.263, 0.233, 0.234, 0.366, 0.245,
    0.085, 0.103, 0.252, 0.018, 0.321])
USGW =  np.array([
    0.184, 0.200, 0.115, -0.258, 0.273, 0.243, 0.167, 0.226, 0.384, 0.231,
    0.090, 0.103, 0.234, -0.017, 0.313])

NOB =  np.array([
    0.487, 0.450, 0.001, -0.610, 0.335, 0.295, 0.001, 0.284, 0.266, 0.245,
    0.383, 0.359, 0.174, 0.033, 0.326])
NOT = np.array([
    0.332, 0.609, -0.126, -0.313, 0.021, 0.063, 0.054, 0.048, 0.146, 0.543,
    0.39, 0.171, 0.141, 0.147, 0.22])
NOGW = np.array([
    0.356, 0.426, 0.142, -0.699, 0.139, 0.171, -0.143, 0.102, 0.276, 0.469,
    0.289, 0.184, 0.125, 0.077, 0.338])
```

```python
market_last[0]
```

```python
years = np.linspace(2005,2019,15)
```

```python
start_value = market_last[0]
```

```python
def value_calc(weights,start_value):
    value_list = []
    for i in range(0,len(weights+1)):
        if i == 0:
            value_list.append(start_value)
        else:
            value_list.append((1+weights[i-1])*value_list[i-1])
    return value_list
```

```python
value_USB = value_calc(USB,start_value)
value_UST = value_calc(UST,start_value)
value_USGW = value_calc(USGW,start_value)

value_NOB = value_calc(NOB,start_value)
value_NOT = value_calc(NOT,start_value)
value_NOGW = value_calc(NOGW,start_value)
```

```python
value_USB
```

```python
years
```

```python
plt.plot(years ,market_last, label='SP500')
plt.plot(years,value_USB,label='US optimal')
plt.plot(years,value_UST,label='US Threshold')
plt.plot(years,value_USGW,label='US Greenwash')
plt.plot(years,value_NOB,label='Nordic optimal')
plt.plot(years,value_NOT,label='Nordic Threshold')
plt.plot(years,value_NOGW,label='Nordic Greenwash')


plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()

# Plot a red dot at the portfolio with the highest Sharpe ratio
plt.show()
```

```python
plt.plot(years ,market_last, label='SP500')
plt.plot(years,value_USB,label='US optimal')
plt.plot(years,value_NOB,label='Nordic optimal')



plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()

# Plot a red dot at the portfolio with the highest Sharpe ratio
plt.savefig('best_vs_SP500.png')
plt.show()
```

```python
plt.plot(years ,market_last, label='SP500')
plt.plot(years,value_UST,label='US Threshold')
plt.plot(years,value_NOT,label='Nordic Threshold')

plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()

plt.savefig('Threshold_vs_SP500.png')
# Plot a red dot at the portfolio with the highest Sharpe ratio
plt.show()
```

```python
plt.plot(years ,market_last, label='SP500')
plt.plot(years,value_USGW,label='US Greenwash')
plt.plot(years,value_NOGW,label='Nordic Greenwash')

plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.savefig('GW_vs_SP500')
```

```python
start_date = '2005-01-01'
end_date = '2020-01-01'
marketdata = yf.download("^GSPC", start=start_date, end=end_date)
mreturns = marketdata['Adj Close']
mreturns.index = pd.to_datetime(mreturns.index)
mreturns
```

```python
mpct = mreturns.ffill().pct_change().resample('y').sum()
mpct
```

```python
pctlst = mpct.to_list()
```

```python
years = np.linspace(2006,2020,15)
```

```python
plt.plot(years ,pctlst, label='SP500')
plt.plot(years,USB,label='US optimal')
plt.plot(years,NOB,label='Nordic optimal')



plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()

# Plot a red dot at the portfolio with the highest Sharpe ratio
plt.savefig('PCT_best')
plt.show()
```

```python
plt.plot(years ,pctlst, label='SP500')
plt.plot(years,UST,label='US Threshold')
plt.plot(years,NOT,label='Nordic Threshold')



plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()

# Plot a red dot at the portfolio with the highest Sharpe ratio
plt.savefig('PCT_threshold')
plt.show()
```

```python
plt.plot(years ,pctlst, label='SP500')
plt.plot(years,USGW,label='US GW')
plt.plot(years,NOGW,label='Nordic GW')



plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()

# Plot a red dot at the portfolio with the highest Sharpe ratio
plt.savefig('PCT_GW')
plt.show()
```
